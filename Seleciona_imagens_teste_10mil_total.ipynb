## Cria o conjunto de teste
### - As imagens a serem usadas não foram empregadas no treinamento ou validação
### - Contém 10 mil imagens ao todo: 5 mil positivas e 5 mil negativas

```python
import pandas as pd
import random
from random import sample
```

```python
# Dataframe contendo todas as linhas do csv que não foram usadas no treinamento
df_completo = pd.read_csv ('test_pos_1500000_final.csv', sep = ';')
df_completo
```

```python
# Lista contendo os StudyInstanceUID únicos
lista_exames_tep_pos = (df_completo[df_completo['negative_exam_for_pe'] == 0]['StudyInstanceUID'].dropna().unique()).tolist()
print ("Total de exames positivos para TEP: ", len (lista_exames_tep_pos))

lista_exames_tep_neg = (df_completo[df_completo['negative_exam_for_pe'] == 1]['StudyInstanceUID'].dropna().unique()).tolist()
print ("Total de exames negativos para TEP: ", len(lista_exames_tep_neg))

print ("Total de exames: ", len (lista_exames_tep_pos)+len (lista_exames_tep_neg))
```

```python
# Seleciona 300 exames positivos e 300 exames negativos
# Cria um dataframe contendo todas as imagens desses exames
lista_300_exames_tep_pos = sample(lista_exames_tep_pos,300)
lista_300_exames_tep_neg = sample(lista_exames_tep_neg,300)

df_300_exames_tep_pos = df_completo[df_completo['StudyInstanceUID'].isin(lista_300_exames_tep_pos)]
df_300_exames_tep_neg = df_completo[df_completo['StudyInstanceUID'].isin(lista_300_exames_tep_neg)]

print ("Total de imagens dos 300 exames positivos: ", len(df_300_exames_tep_pos))
print ("Total de imagens dos 300 exames negativo: ", len(df_300_exames_tep_neg))
```

```python
df_imagens_tep_pos = (df_300_exames_tep_pos[df_300_exames_tep_pos['pe_present_on_image'] == 1])
print ("Total de imagens positivas dos 300 exames positivos: ", len(df_imagens_tep_pos))
```

```python
# Seleciona 5000 imagens positivas de exames positivos
# Seleciona 5000 imagens negativas de exames negativos
# Cria csv com as informações dessas imagens
random.seed(5)

df_imagens_tep_pos = (df_300_exames_tep_pos[df_300_exames_tep_pos['pe_present_on_image'] == 1])
df_imagens_tep_neg = (df_300_exames_tep_neg[df_300_exames_tep_neg['pe_present_on_image'] == 0])

teste1 = df_imagens_tep_pos.sample(n = 5000, replace = False)
teste2 = df_imagens_tep_neg.sample(n = 5000, replace = False)

bigdata = pd.concat([teste1, teste2], ignore_index=True, sort=True)

bigdata.to_csv('imagens_teste_10000_total.csv', index=False)

print ("Total de imagens do CSV gerado:", len(bigdata))
```
