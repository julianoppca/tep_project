# Manipulação de Arquivos
#### 1º) Acessa o csv contendo 1,5 milhão de imagens do dataset RSNA
#### 2º) Seleciona 1.500 exames positivos e 1.500 negativos para TEP
#### 3º) Seleciona 50.000 imagens positivas e 50.000 imagens negativas para TEP
#### 4º) Cria arquivo .csv contendo as informações dessas 100.000 imagens sorteadas

```python
import pandas as pd
import random
from random import sample
```

```python
# Dataframe contendo todas as linhas do csv de 1.5 milhão de imagens
df_completo = pd.read_csv ('train_1500000_final_novo.csv')
df_completo
```

```python
# Lista contendo os StudyInstanceUID únicos
lista_exames_tep_pos = (df_completo[df_completo['negative_exam_for_pe'] == 0]['StudyInstanceUID'].dropna().unique()).tolist()
print ("Total de exames positivos para TEP: ", len (lista_exames_tep_pos))

lista_exames_tep_neg = (df_completo[df_completo['negative_exam_for_pe'] == 1]['StudyInstanceUID'].dropna().unique()).tolist()
print ("Total de exames negativos para TEP: ", len(lista_exames_tep_neg))

print ("Total de exames: ", len (lista_exames_tep_pos)+len (lista_exames_tep_neg))
```

```python
# Seleciona 1500 exames positivos e 1500 exames negativos
# Cria um dataframe contendo todas as imagens desses exames
lista_1500_exames_tep_pos = sample(lista_exames_tep_pos,1500)
lista_1500_exames_tep_neg = sample(lista_exames_tep_neg,1500)

df_1500_exames_tep_pos = df_completo[df_completo['StudyInstanceUID'].isin(lista_1500_exames_tep_pos)]
df_1500_exames_tep_neg = df_completo[df_completo['StudyInstanceUID'].isin(lista_1500_exames_tep_neg)]

print ("Total de imagens dos 1500 exames positivos: ", len(df_1500_exames_tep_pos))
print ("Total de imagens dos 1500 exames negativo: ", len(df_1500_exames_tep_neg))
```

```python
# Seleciona 50000 imagens positivas de exames positivos
# Seleciona 50000 imagens negativas de exames negativos
# Cria csv com as informações dessas imagens
random.seed(5)

df_imagens_tep_pos = (df_1500_exames_tep_pos[df_1500_exames_tep_pos['pe_present_on_image'] == 1])
df_imagens_tep_neg = (df_1500_exames_tep_neg[df_1500_exames_tep_neg['pe_present_on_image'] == 0])

teste1 = df_imagens_tep_pos.sample(n = 50000, replace = False)
teste2 = df_imagens_tep_neg.sample(n = 50000, replace = False)

bigdata = pd.concat([teste1, teste2], ignore_index=True, sort=True)

bigdata.to_csv('split_50_50_junho/05_train_imagens_50000.csv', index=False)

print ("Total de imagens do CSV gerado:", len(bigdata))
```
